## Importing Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

## Importing Dataset

```python
df = pd.read_csv('daily-min-temperatures.csv')

df.head()
```

```python
df = df.set_index('Date')

df.head()
```

```python
df.plot(figsize =(10,6))
plt.show()
```

## Check for Autocorrelation

There is a quick, visual check that we can do to see if there is an autocorrelation in our time series dataset.

Pandas provides a built-in plot to do this, called the lag_plot() function.

```python
from pandas.plotting import lag_plot

lag_plot(df)
plt.show()
```

We can see a large ball of observations along a diagonal line of the plot. 

It clearly shows a relationship or some correlation.

## Autocorrelation Plots

We can plot the correlation coefficient for each lag variable.

This can very quickly give an idea of which lag variables may be good candidates for use in a predictive model and how the relationship between the observation and its historic values changes over time.

Pandas provides a built-in plot called the autocorrelation_plot() function for this.

```python
from pandas.plotting import autocorrelation_plot

autocorrelation_plot(df)
plt.show()
```

It shows the swing in positive and negative correlation as the temperature values change across summer and winter seasons each previous year.

## ACF

The statsmodels library also provides a version of the plot in the plot_acf() function as a line plot.

```python
from statsmodels.graphics.tsaplots import plot_acf

plot_acf(df, lags=31)    ## limiting the lag variables evaluated to 31 for readability.
plt.show()
```

## PACF

```python
from statsmodels.graphics.tsaplots import plot_pacf

plot_pacf(df, lags=50)
plt.show()
```
