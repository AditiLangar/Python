### Outlier Treatment

```python
for col in df.iloc[:, :-1].columns:
    if df[col].dtype != object:
        
        q1, q3 = np.percentile(df[col], [25,75])
        
        iqr = q3-q1
        lower_fence = q1 - (1.5 * iqr)
        upper_fence = q3 + (1.5 * iqr)
        
        df[col] = df[col].apply(lambda x: upper_fence if x > upper_fence else lower_fence if x< lower_fence else x)
        sns.boxplot(x = df[col], data= df, palette='rainbow', orient='h')
        plt.show()
        print('*'*100, '\n'*2)
```

## Scaling the variables

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler
sc = StandardScaler()
min_max = MinMaxScaler()
```

### Splitting Data 

```python
X = df.drop(['price'], axis = 1)
y = df['price'] # Target

X.head()
```

```python
X = min_max.fit_transform(X)
X[0]
```

## Checking VIF Scores

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
h = [variance_inflation_factor(X, i) for i in range(X.shape[1])]
print(h)

pd.DataFrame(h, index= df.iloc[:,:-1].columns, columns = ['VIF Score'])
```

As expected, there are a large number of variables with high VIF, but leaving them as such else we will be left with a very few variables

## Model Making

### train_test_split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, 
                                                    y, 
                                                    test_size = 0.3, 
                                                    random_state = 42)
```

### Decision Tree Regressor

```python
# importing the regressor 

from sklearn.tree import DecisionTreeRegressor  
```

```python
??DecisionTreeRegressor
```

```python
# creating a regressor object
# Using default parameters for now

model = DecisionTreeRegressor(random_state = 42)  
```

```python
model.fit(X_train,y_train)

model
```

```python
i = model.feature_importances_

pd.DataFrame(i, index= boston.feature_names, columns = ['Feature Importance'])
```

```python
## Checking depth of the model

model.get_depth()
```

## Plotting Decision Tree

```python
from sklearn import tree
```

```python
tree.plot_tree(model, feature_names = boston.feature_names)
```

```python
y_pred = model.predict(X_test)

y_pred[:15]
```

### Checking Accuracy Metrices

```python
from sklearn import metrics
from sklearn.metrics import accuracy_score, classification_report,confusion_matrix, mean_absolute_error, mean_squared_error, r2_score


## checking reports

print('MAE: ',mean_absolute_error(y_test, y_pred))

print('MSE: ', mean_squared_error(y_test, y_pred))

print('RMSE: ',np.sqrt(mean_squared_error(y_test, y_pred)))

print('R2: ',r2_score(y_test, y_pred)*100)

print('Adjusted R^2:',1 - (1-metrics.r2_score(y_test, y_pred))*(len(y_test)-1)/(len(y_test)-X_test.shape[1]-1))
```
